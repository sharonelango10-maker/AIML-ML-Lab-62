```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, classification_report
```

```python
#Load Real dataset 
data = load_breast_cancer()

X = data.data
Y = data.target

print("Classes:", data.target_names)
#malignant,benign

#Split data 
X_train, X_test, Y_train, Y_test, = train_test_split(X, Y, test_size=0.3)
```

```python
#-------------------------------------------------
#Logistic_Regression
#-------------------------------------------------
lr = LogisticRegression(max_iter=5000)
lr.fit(X_train, Y_train)
pred_lr = lr.predict(X_test)
print("Logistic TestAccuracy:", accuracy_score(Y_test, pred_lr))
pred_lr_train = lr.predict(X_train)
print("Logistic TrainAccuracy:", accuracy_score(Y_train, pred_lr_train))
```

```python
#-------------------------------------------------
#Decision_Tree
#-------------------------------------------------
dt = DecisionTreeClassifier()
dt.fit(X_train, Y_train)
pred_dt = dt.predict(X_test)
print("Decision Tree TestAccuracy:", accuracy_score(Y_test, pred_dt))
pred_dt_train = dt.predict(X_train)
print("Decision Tree TrainAccuracy:", accuracy_score(Y_train, pred_dt_train))
```

```python
#-------------------------------------------------
#K Nearest Neighbors
#-------------------------------------------------
KNN =  KNeighborsClassifier()
KNN.fit(X_train, Y_train)
pred_KNN = KNN.predict(X_test)
print("KNN TestAccuracy:", accuracy_score(Y_test, pred_KNN))
pred_KNN_train = KNN.predict(X_train)
print("KNN TrainAccuracy:", accuracy_score(Y_train, pred_KNN_train))
```

```python
#-------------------------------------------------
#Gaussian NB
#-------------------------------------------------
GNB =  GaussianNB()
GNB.fit(X_train, Y_train)
pred_GNB = GNB.predict(X_test)
print("GNB TestAccuracy:", accuracy_score(Y_test, pred_GNB))
pred_GNB_train = GNB.predict(X_train)
print("GNB TrainAccuracy:", accuracy_score(Y_train, pred_GNB_train))
```

```python
#--------------------------------------------------
#MLPClassifier
#--------------------------------------------------
MLP =  MLPClassifier(max_iter=2000)
MLP.fit(X_train, Y_train)
pred_MLP = MLP.predict(X_test)
print("MLP TestAccuracy:", accuracy_score(Y_test, pred_MLP))
pred_MLP_train = MLP.predict(X_train)
print("MLP TrainAccuracy:", accuracy_score(Y_train, pred_MLP_train))
```
