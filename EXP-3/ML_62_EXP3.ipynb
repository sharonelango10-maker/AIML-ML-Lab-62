```python
# Step 1 : Import Libraries 
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Step 2 : Input data 
X = np.array([1,2,3]).reshape(-1,1) #input 
Y = np.array([2,3,5]) #output 
```

```python
# Step 3 : Create model anf fit 
model = LinearRegression()
model.fit(X,Y)
```

```python
# Step 4 : Get coefficients (MLE Estimation)
w = model.coef_[0]
b = model.intercept_

print(f"w (slope) = {w}")
print(f"b (intercept) = {b}")
```

```python
# Step 6 : Make predictions 
Y_pred = model.predict(X)
print("\nPredictions for training data:")
for xi, yi, ypi in zip(X.flatten(), Y, Y_pred):
    print(f"X={xi}, Actual Sales={yi}, Predicted Sales={ypi:.2f}")

# Step 7 : Calculate metrics 
mse = mean_squared_error(Y,Y_pred)
r2 = r2_score(Y,Y_pred)

print(f"\nMean Squared Error (MSE) = {mse:.4f}")
print(f"R² Score = {r2:.4f}")
```

# Maximum Likelihood Estimation (MLE)

```python
from scipy.optimize import minimize 

#Data 
X = np.array([1,2,3]) #input 
Y = np.array([2,3,5]) #output
```

```python
#Negation log_likelihood estimation 
def neg_log_likelihood(params):
    w, b = params 
    sigma2 = 1 #assume variance = 1 for simplicity 
    Y_pred = w*X + b
    nll = 0.5*np.sum((Y - Y_pred)**2 / sigma2) # - log-likelihood up to constant 
    return nll 

#initial guess for w , b
initial_guess = [0,0]

#Minimize negative log-likelihood 
result = minimize(neg_log_likelihood, initial_guess)
w_mle, b_mle = result.x
print(f"MLE w = {w_mle}")
print(f"MLE b = {b_mle}")
```

```python
#Step 6 : Make Predictions 
#Make predictions using estimated parameters 
Y_pred = w_mle * X +b_mle

print ("\nPredictions for training data:")
for xi, yi, ypi in zip (X, Y, Y_pred):
    print(f" X={xi}, Actual ={yi}, Predicted={ypi:.2f}") 
```

```python
mse=mean_squared_error(Y,Y_pred)
r2=r2_score(Y,Y_pred)
print(f"\nMean squared error(MSE)={mse:.4f}")
print(f"R2 Score={r2:.4f}")
```

```python
# -------------------------------
from matplotlib import pyplot as plt
#Data 
X = np.array([1,2,3]) #input 
Y = np.array([2,3,5]) #output
n=len(X)

#---------------------------------
#Loss function J(w1)
#---------------------------------
def loss(w1):
    w0 = np.mean(Y) - w1*np.mean(X)
    Y_pred = w1 * X + w0
    return np.sum((Y -Y_pred)**2)

#---------------------------------
#Gradient of J w.r.t w1
#---------------------------------
def gradient(w1):
    w0 = np.mean(Y) - w1*np.mean(X)
    Y_pred = w1 * X + w0
    return -2 * np.sum(X*(Y -Y_pred))

#---------------------------------
#Gradient Descent 
#---------------------------------
lr = 0.1
w1 = 4          # initial guess (intentionally bad)
iterations = 15

w1_values = []
loss_values = []

for i in range (iterations):
    w1_values.append(w1)
    loss_values.append(loss(w1))

    grad = gradient(w1)
    w1 = w1-lr*grad 

#---------------------------------
# Plot Loss Curve 
#---------------------------------
w_space = np.linspace(-2,4,200)
loss_space = [loss(w) for w in w_space]

plt.figure(figsize=(8,5))
plt.plot(w_space, loss_space, label="J(w1)")
plt.scatter(w1_values, loss_values, color="red", label="GD steps")
plt.xlabel("w1 (slope)")
plt.ylabel ("Sum  of Squared Errors (Loss)")
plt.title ("Gradient Descent on Slope Only (w1)")
plt.legend()
plt.grid(True)
plt.show()
```
